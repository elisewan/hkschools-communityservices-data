# School Email Finder - Brave Search

Finds contact emails for schools using the Brave Search API.

```
import pandas as pd
import requests
import time
import io
import os
from google.colab import files
from IPython.display import display, HTML

# --- CONFIGURATION ---
# Get your free key: https://api.search.brave.com/register
# NEVER hardcode your API key! Use environment variables.
BRAVE_API_KEY = os.environ.get("BRAVE_API_KEY")

if not BRAVE_API_KEY:
    print("WARNING: BRAVE_API_KEY not found in environment variables.")
    print("Please set it using: os.environ['BRAVE_API_KEY'] = 'your-key'")
    # For quick testing in Colab, you can uncomment and set it here:
    # BRAVE_API_KEY = "your-key-here"  # Remove this line before publishing!

def get_first_link(school_name):
    """Searches Brave and returns the first result's URL."""
    url = "https://api.search.brave.com/res/v1/web/search"
    headers = {
        "Accept": "application/json",
        "X-Subscription-Token": BRAVE_API_KEY
    }
    params = {"q": f"{school_name} site:schooland.hk", "count": 1}

    try:
        response = requests.get(url, headers=headers, params=params, timeout=10)
        if response.status_code == 200:
            results = response.json().get('web', {}).get('results', [])
            return results[0]['url'] if results else "No Link Found"
        return f"Error: {response.status_code}"
    except:
        return "Connection Error"

# Upload file
print("Please upload your Excel file with school names:")
uploaded = files.upload()

if uploaded:
    fn = list(uploaded.keys())[0]
    df_in = pd.read_excel(io.BytesIO(uploaded[fn]), header=None)
    school_names = df_in.iloc[:, 0].dropna().tolist()

    all_data = []
    print(f"\n Fetching top results for {len(school_names)} schools...\n")

    for i, name in enumerate(school_names):
        name_str = str(name).strip()
        link = get_first_link(name_str)
        all_data.append({"School Name": name_str, "First Result Link": link})

        # Live display
        df_display = pd.DataFrame(all_data)
        from google.colab import output
        output.clear(wait=True)
        print(f"Progress: {i+1}/{len(school_names)}")
        display(HTML(df_display.to_html(render_links=True, escape=False, index=False)))
        time.sleep(1.1)  # Rate limit

    # Save and download
    final_df = pd.DataFrame(all_data)
    final_df.to_excel("School_First_Links.xlsx", index=False)
    files.download("School_First_Links.xlsx")
    print("\n Process complete!")
```
